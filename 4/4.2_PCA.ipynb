```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.feature_extraction import DictVectorizer
```

```python
close_price_df = pd.read_csv('close_prices.csv')
close_price_df.head()
```

```python
encoder = DictVectorizer(sparse = False)
data = [close_price_df['date']]
data = pd.concat(data, axis=1) 

encoded_data = encoder.fit_transform(data.T.to_dict().values())

cat_oh = pd.DataFrame(data=encoded_data)
print(cat_oh.head())
```

```python
close_price_df_clear = close_price_df.drop(['date'], axis = 1)
```

```python
close_price_df_clear.head()
```

```python
# обучаем метод главных компонент
pca = PCA(n_components=10)
pca.fit(close_price_df_clear)
```

```python
# Скольких компонент хватит, чтобы объяснить 90% дисперсии
count = 0
summary = 0
for i in range(len(pca.explained_variance_ratio_)):
    count+=1
    summary  += pca.explained_variance_ratio_[i]
    if summary >= 0.9:
        break
print("Need components %d" % count)
```

```python
djia_index_df = pd.read_csv('djia_index.csv')
djia_index_df.head()
```

```python
comp = pca.transform(close_price_df_clear)
```

```python
djia_index_df['^DJI'].shape
```

```python
comp.T[0].shape
```

```python
# Чему равна корреляция Пирсона между первой компонентой и индексом Доу-Джонса? 
cor_mat = np.corrcoef(comp.T[0], djia_index_df['^DJI'])
print(round(cor_mat[1][0], 2))
```

```python
# найти компанию, которая имеет наибольший вес в первой компоненте
brand_index =np.argmax(pca.components_[0])
print(close_price_df_clear.columns[brand_index])
```
